```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

jute = "data/jute.csv"
wheat = "data/wheat.csv"

dataset = pd.read_csv(wheat)
```

```python
dataset
```

```python
X = dataset[['Area of Land in acres','Rainfall in MM','Year']]
# X = dataset[['Area Name']]
X = X.values
#X = X.reshape(-1,2)
y = dataset[['Amount of Production in Tons']]
y = y.values
```

```python
%matplotlib notebook
plt.title("Amount of Rice Production in Tons in our Dataset")
plt.xlabel("DATA X: Area of Land in acres' , ' Rainfall in MM', 'Year")
plt.ylabel("DATA Y : Amount of Production in Tons")
plt.plot(X,y,"+") 
plt.show()
```
