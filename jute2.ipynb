```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from sklearn import linear_model
from sklearn.linear_model import Ridge

from sklearn.svm import LinearSVC
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import mean_squared_error

#from sklearn.neural_network import MLPClassifier

dataset = pd.read_csv("data/jute.csv")
# dataset = pd.read_csv("just wheat featured scaled.csv")
#dataset = pd.read_csv("new temp.csv")
categorical = pd.read_csv("data/categorical.csv")
```

```python
choice = 1

if choice == 1:
    #X = dataset[['Area of Land in acres','Rainfall in MM','Year']]
    #X = dataset[['Area of Land in acres','Rainfall in MM', 'Temparature degree cellcius']]
    X = dataset[['Area of Land in acres','Rainfall in MM']]
    y = dataset[['Amount of Production in Tons']]
    
elif choice == 2:
    
    X = dataset[['Area of Land in acres','Amount of Production in Tons','Year']]
    y = dataset[['Rainfall in MM']]
    # y=lab_enc.fit_transform(y)
    
    
elif choice == 3:
    X = dataset[['Year']]
    y = dataset[['Area of Land in acres','Amount of Production in Tons','Rainfall in MM']]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=20)
#X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3,random_state=1)
```

```python
selectModel = 2

if selectModel == 1:
    clf = linear_model.LinearRegression()
    
elif selectModel == 2:
    clf = RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1)
    
elif selectModel == 3:
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
    
elif selectModel == 4:
    clf = SVC(gamma=0.001, C=100.)
    
elif selectModel == 5:
    clf = Ridge(alpha=1.0, copy_X=True, fit_intercept=True, max_iter=None,normalize=False, random_state=None, solver='auto', tol=0.001)
    
elif selectModel == 6:
    clf = linear_model.Lasso(alpha=0.1, copy_X=True, fit_intercept=True, max_iter=1000,
   normalize=False, positive=False, precompute=False, random_state=None,
   selection='cyclic', tol=0.0001, warm_start=False)
    
elif selectModel == 7:
    clf = AdaBoostClassifier()
    
elif selectModel == 8:  
    #dataset = columns ['Crop Name' ,'Area Name','Area of Land in acres','Amount of Production in Tons','Rainfall in MM','Year']
    clf = KNeighborsClassifier(n_neighbors = 1.2)

elif selectModel == 9:
    clf = LinearSVC(random_state=0)
    
y_train = y_train.values.ravel()
model = clf.fit(X_train, y_train)
```

```python
accuracy= clf.score(X_test, y_test)
print("The accuracy of the network is:", accuracy)
print("Prediction for the Test set is:")
print( clf.predict(X_test) )
print("Lets Compare it to the original target test set:")
print(y_test)
```
